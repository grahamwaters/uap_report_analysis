```python
# UAP Colors
import re
import statistics
from tqdm import tqdm

def colorizer(df):
    # create the comments_colors column and initialize it to "None"
    df['comments_colors'] = 'None'
    df['comments_modecolor'] = 'None'
    df['comments_clean'] = df['comments_clean'].astype(str) # to address a type error that was occuring
    # create a list of colors
    colors = ['red', 'orange', 'yellow', 'green', 'blue', 'purple', 'pink', 'brown', 'black', 'white']
    # create a regex pattern to match colors
    pattern = re.compile(r'\b(' + '|'.join(colors) + r')\b')
    # iterate over the rows in the dataframe
    for i, row in tqdm(df.iterrows(), total = df.shape[0]):
        # get the comments_clean column
        comments_clean = row['comments_clean']
        # find the colors in the comments_clean column
        color_list = pattern.findall(comments_clean)
        colors_string = str(color_list)
        if color_list:
            # get the mode of the colors found
            mode_color = statistics.mode(color_list)
            # add the list of colors to the comments_colors column as a string
            df.loc[i, 'comments_colors'] = colors_string
            # add mode color to the comments_modecolor column
            df.loc[i, 'comments_modecolor'] = str(mode_color)
    return df
           
```

```python
# apply the function to the dataframe
df = colorizer(df)
```
