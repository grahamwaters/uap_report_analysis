```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read in the data from the intermediary folder
df = pd.read_csv('../data/intermediary/ufos_clean.csv', low_memory=False)
# strip whitespace from column names
df.columns = df.columns.str.strip()
df.head()
```

```python
# Initialize a count variable
count = 0

# Iterate through the rows in the df
for i, row in df.iterrows():
    # Try to convert the row to a float
    try:
        float(row['latitude'])
    except ValueError:
        # If the conversion fails, increment the count
        count += 1
        # drop that row
        df.drop(i, inplace=True)
# Print the count
print(count)
```

```python
# Initialize a count variable
count = 0

# Iterate through the rows in the df
for i, row in df.iterrows():
    # Try to convert the row to a float
    try:
        float(row['longitude'])
    except ValueError:
        # If the conversion fails, increment the count
        count += 1
        # drop that row
        df.drop(i, inplace=True)
# Print the count
print(count)
```

```python
from datetime import datetime
# Create new columns for year, month, day, hour, minute, and second
df['datetime'] = pd.to_datetime(df['datetime'])
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day'] = df['datetime'].dt.day
df['hour'] = df['datetime'].dt.hour
df['minute'] = df['datetime'].dt.minute
df['second'] = df['datetime'].dt.second

# Create a new column for the day of the week
df['date_posted'] = pd.to_datetime(df['date posted'])
df['time_since_event'] = df['date_posted'] - df['datetime']
df['comment_length_words'] = df['comments'].apply(lambda x: len(x.split()))
df['comment_length_characters'] = df['comments'].apply(lambda x: len(x))
df['comment_unique_words'] = df['comments'].apply(lambda x: len(set(x.split())))

# convert the lat lng to numeric
df['latitude'] = pd.to_numeric(df['latitude'])
df['longitude'] = pd.to_numeric(df['longitude'])



df.info()
```

# The flesch-kincaid grade level of the text

```python
import textstat
#note: this may need to run before we cleaned the text of the comments. For now I'm just going to use the cleaned text. 

def calculate_flesch_kincaid_score(text: str):
    return textstat.flesch_kincaid_grade(text)

df['flesch_kincaid_score'] = df['comments'].apply(calculate_flesch_kincaid_score) # apply the function to the comments column
```

```python
# replace all NaN values in comments_shapes with 'None'
df['comments_shapes'].fillna('None', inplace=True) # replace all NaN values in comments_shapes with 'None'
```
